```python
import pickle
import pandas as pd
```

```python
with open("data.bin", "rb") as f: # "rb" because we want to read in binary mode
    state = pickle.load(f)
```

```python
state
```

```python
from io import StringIO

s=str(state,'utf-8')

data = StringIO(s) 

df=pd.read_csv(data, header = 2, skipfooter=1, engine='python')
```
