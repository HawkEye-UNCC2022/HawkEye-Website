```python
import sqlite3
import pandas as pd
```

# Saving .csv to SQLite database

```python
# reading .csv file saved in local and saving it as dataframe (alpha)
alpha = pd.read_csv('Alpha Roster.csv', header = 2, skipfooter=1)
alpha.head()
```

```python
# formatting SSAN as it will be used as unique identifier in future quering
alpha['SSAN'] = alpha.SSAN.str.replace('-', '')
```

```python
# creating database file and connection instance
conn = sqlite3.connect('airforce.sqlite')
```

```python
# saving dataframe to sqlite db as a table
alpha.to_sql(name='alpha', con=conn, if_exists='replace', index=False)
conn.commit()
```

# Reading saved tables from SQLite database

```python
# reading table from saved sqlite
df = pd.read_sql_query("SELECT * from alpha", conn)
df
```

```python
# saving and checking if .csv file converted to sql table and back is formatted correctly
df.to_csv('output.csv')
```
